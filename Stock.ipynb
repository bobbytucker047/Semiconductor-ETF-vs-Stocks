```python
import yfinance as yf
import matplotlib.pyplot as plt
import pandas as pd
from datetime import datetime



etfs = ['SMH', 'TSM', 'NVDA', 'AMD'] # my current investments in taxable


start_date = '2025-01-01'
end_date = datetime.now().strftime('%Y-%m-%d')  #  Fetching for today's date

print(f"Fetching data for {', '.join(etfs)}...")
print(f"Period: {start_date} to {end_date} (Today)\n")


data = yf.download(etfs, start=start_date, end=end_date, progress=False)


closing_prices = data['Close']


plt.figure(figsize=(12, 7))


for etf in etfs:
    plt.plot(closing_prices.index, closing_prices[etf], label=etf, linewidth=2)


plt.title(f'Bobby Tucker ETF Performance - YTD 2025 (through {datetime.now().strftime("%B %d, %Y")})',
          fontsize=16, fontweight='bold')
plt.xlabel('Date', fontsize=12)
plt.ylabel('Price ($)', fontsize=12)
plt.legend(fontsize=11, loc='best')
plt.grid(True, alpha=0.3)
plt.tight_layout()


plt.show()


print("\n=== YTD 2025 Performance Summary ===")
print(f"Data through: {datetime.now().strftime('%B %d, %Y')}\n")

for etf in etfs:
    start_price = closing_prices[etf].iloc[0]
    end_price = closing_prices[etf].iloc[-1]
    change = end_price - start_price
    pct_change = (change / start_price) * 100

    print(f"{etf}: {pct_change:+.2f}%")
    print(f"  Start: ${start_price:.2f}  Current: ${end_price:.2f}")
    print()
```
